```python
newdata=data.iloc[:,4:6]
display (newdata)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(newdata)
```

```python
display (kmeans.cluster_centers_)
```

```python
print (kmeans.labels_)
```

```python
pd.DataFrame ((kmeans.labels_)).head(50)
```

```python
cluster = pd.concat([data,pd.DataFrame(kmeans.labels_, columns = ['Cluster Number'])],axis=1)
cluster
```

```python
print (len(kmeans.labels_))
```

```python
cluster.to_excel ('29_Jan_moviecluster.xlsx')
display (cluster)
```

```python
print (type(kmeans.labels_))
unique, counts = np.unique(kmeans.labels_, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
newdata['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot(x= 'director_facebook_likes',y = 'actor_3_facebook_likes',data=newdata, hue='cluster', palette='coolwarm',height=10,aspect=1,fit_reg=False)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8)
kmeans.fit(newdata)
unique, counts = np.unique(kmeans.labels_, return_counts=True)
print(dict(zip(unique, counts)))
newdata['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot(x= 'director_facebook_likes',y = 'actor_3_facebook_likes',data=newdata, hue='cluster',
           palette='coolwarm',height=10,aspect=1,fit_reg=False)
plt.show()
```
